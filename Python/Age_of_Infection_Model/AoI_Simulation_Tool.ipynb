# **The Age of Infection Model**  
Here, we address the integro-differential equation representing the Kermack and McKendrick **Age-of-Infection (AoI)** epidemic model 
\begin{equation}\tag{1}
    \displaystyle S^{\prime}(t)=\beta_0 S(t)\int_{0}^{\infty}[S^{\prime}(t-\tau)]A(\tau)\ d\tau,
\end{equation}
where $S(t)$ denotes the number of susceptibles at time $t$ and $\beta_0>0$ represents the rate of effective contacts. Furthermore, $A(\tau)=\pi(\tau)B(\tau)\in L^1(\mathbb{R}^+_0)$ is the  mean infectivity of members of the population with infection age $\tau,$ while $B(\tau)$ denotes the fraction of infected members remaining infected at infection age $\tau$ and $0\leq\pi(\tau)\leq 1$  is the mean infectivity at infection age $\tau.$ 

It is assumed that the disease outbreak begins at time $t=0,$ so that $S(t)=N,$ for $t<0,$ and that there are no disease deaths, so that the total population size is a constant $N$. If we introduce the total infectivity $\varphi(t)=-\int_{0}^{+\infty}{A(s)S^{\prime}(t-s)ds},$ at time $t,$ equation (1) can be rewritten as follows 
\begin{equation}\tag{2}
    \begin{aligned}
        S^{\prime}(t)&=-\beta_0 S(t)\varphi(t)\\
        \varphi(t)&=\varphi_0(t)+ \beta_0 \int_{0}^{t}A(t-\tau) S(\tau)\varphi(\tau) \ d\tau,
    \end{aligned}
\end{equation}
where the function $\varphi_0(t)=-\int_{t}^{\infty}A(\tau)S^{\prime}(t-\tau) \ d\tau$ is the total infectivity, at time $t,$ of members of the population who were infected at  $t=0.$
We refer to [[Brauer_Castillo-Chavez_Feng_2019]](https://doi.org/10.1007/978-1-4939-9828-9), [[Brauer_2005]](https://doi.org/10.1016/j.mbs.2005.07.006) and references therein for a detailed description and analysis of AoI epidemic models of the form (1) and (2).

# **The NSFD Scheme**  
The present code is based on the **Non-Standard Finite Difference (NSFD)** method proposed in [[Messina_Pezzella_Vecchio_2022]](https://doi.org/10.3934/jcd.2021029). Let $t_n=nh,$  where $n=0,1,\dots,$ denote a uniform mesh with stepsize $h>0.$ The NSFD scheme then reads
\begin{equation}\tag{3}
    \begin{aligned}
        &S_{n+1}=S_n-h \beta_0 S_{n+1}\varphi_n \\
        &\varphi_{n+1}=\varphi_0(t_{n+1})+h \beta_0\sum_{j=0}^{n} A(t_{n+1-j})S_{j+1}\varphi_j,
    \end{aligned}
\end{equation}
for $n=0,1,\ldots,$ where $S_0=S(0)>0$, $\varphi_0=\varphi_0(0)\geq0$ and $S_n\approx S(t_n),$ $\varphi_n\approx \varphi(t_n).$ 

A rigorous theoretical analysis of the numerical properties and of the dynamical consistency of the method (3) is provided in [[Messina_Pezzella_Vecchio_2022]](https://doi.org/10.3934/jcd.2021029). We briefly summarize the principal outcomes:
- The NSFD scheme (3) is consistent and convergent of order 1. Hence, the approximation error $e(h;t_n) = \left\| \left[ \begin{array}{c} S(t_n) \\ \varphi(t_n) \end{array} \right] -
\left[ \begin{array}{c} S_n \\ \varphi_n \end{array} \right] \right\|$ vanishes linearly with $h.$
- The method (3) is unconditionally non-negative. More specifically, independently of $h>0$, $S_n>0$ and $\varphi_n\geq0$ for all $n\in\mathbb{N}.$ Furthermore, the NSFD solution is coherently bounded since $0<S_\infty<S_n<S_0$ and $0\leq\varphi_n\leq N\sup_{t\geq0}A(t),$ for all $n\in\mathbb{N}.$
- The discretization (3) is monotonicity-preserving since, for any $h>0$ the sequence $\{S_n\}_{n \in \mathbb{N}}$ is  non-increasing.  Consequently, the limit $S_\infty(h)=\lim_{n\to \infty} S_n,$ exists and is referred to as the *numerical final size*.
- Let $R_0(h)=h \beta_0 N\sum_{n=0}^{+\infty}A(t_{n+1})$ denote the *numerical basic reproduction number*.  A discrete *invasion criterior* holds true, i.e. the NSFD solution initially grows exponentially as $(1+rh)^n$ with $r>0$ iff $R_0(h)>1.$
- The numerical final size satisfies $\lim\limits_{h\to 0}\left(\log\left({\frac{S_0}{S_\infty(h)} }\right)-h \beta_0(N-S_\infty(h))\sum_{n=0}^{\infty}A(t_{n+1}) \right)=0,$ which is a direct discrete counterpart of the continuous-time final size relation (see, for instance, [[Brauer_2008]](https://doi.org/10.3934/mbe.2008.5.681))

```python
%gui tk
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import numpy as np
import matplotlib
matplotlib.use("TkAgg")
import matplotlib.pyplot as plt
from NSFD_AoI import NSFD_AoI

#====================|
# Default Parameters |
#====================|
DEFAULTS = {
    'S0': 990.0, 'N': 1000.0, 'T': 10.0, 'h': 0.1, 'beta': 2.5e-3,
    'A(t)': 'exp(-3.0*t)', 'Φ₀(t)': None
}
# Global storage for simulation results
simulation_results = {'t': None, 'Y': None, 'P': None}

#====================|
# Main GUI Functions |
#====================|
def run_simulation():
    """
    Run the Age-of-Infection simulation using the NSFD_AoI solver.
    Dynamically creates plots for S(t), φ(t), A(t), and Φ₀(t) if provided.
    Results are stored globally for possible export.
    """
    try:
        # Retrieve user inputs
        N = safe_float(entries['N'].get())
        S0 = safe_float(entries['S0'].get())
        beta = safe_float(entries['beta'].get())
        T = safe_float(entries['T'].get())
        h = safe_float(entries['h'].get())
        A_expr = entries['A(t)'].get()
        phi0_expr = entries['Φ₀(t)'].get()
        # Convert expressions to callable functions
        A = make_function(A_expr)
        phi0 = make_function(phi0_expr)
        # Build problem dictionary
        AoI_Problem = {
            'N': N, 'S0': S0, 'beta': beta, 'T': T,
            'h': h, 'A': A, 'phi0': phi0, 'verbosity': 1
        }
        # Run simulation
        t_AoI, Y_AoI, P_AoI = NSFD_AoI(problem=AoI_Problem)
        # Store results globally
        simulation_results['t'] = t_AoI
        simulation_results['Y'] = Y_AoI
        simulation_results['P'] = P_AoI
        # Prepare plots dynamically
        plots = [
            {'y': Y_AoI[0, :], 'color': 'c', 'title': 'Susceptibles S(t)', 'ylabel': 'S(t)'},
            {'y': Y_AoI[1, :], 'color': 'g', 'title': 'Mean Infectivity φ(t)', 'ylabel': 'φ(t)'}
        ]
        if A is not None:
            plots.append({'y': A(t_AoI), 'color': 'b', 'title': 'Kernel A(t)', 'ylabel': 'A(t)'})
        if phi0 is not None:
            plots.append({'y': phi0(t_AoI), 'color': 'm', 'title': 'Initial Infectivity Φ₀(t)', 'ylabel': 'Φ₀(t)'})
        # Create figure
        n_plots = len(plots)
        n_cols = 2
        n_rows = (n_plots + 1) // 2
        fig, axs = plt.subplots(n_rows, n_cols, figsize=(6*n_cols, 4*n_rows))
        axs = np.array(axs).reshape(-1)
        for i, p in enumerate(plots):
            plot_on_ax(axs[i], t_AoI, p['y'], p['color'], p['title'], ylabel=p['ylabel'])
        # Hide unused subplots
        for j in range(i + 1, len(axs)):
            axs[j].axis('off')
        plt.tight_layout()
        plt.show(block=True)
    except Exception as e:
        messagebox.showerror("Simulation Error", f"An error occurred:\n{str(e)}")


def save_results():
    """
    Save simulation results (t_AoI, Y_AoI, P_AoI) to file.
    User can select format (.csv, .npz, .txt) from dropdown menu.
    """
    if simulation_results['t'] is None:
        messagebox.showwarning("No Data", "Please run a simulation before saving results.")
        return
    file_format = save_format_var.get()
    filetypes = [(f"{file_format.upper()} file", f"*.{file_format}")]
    filepath = filedialog.asksaveasfilename(defaultextension=f".{file_format}",
                                            filetypes=filetypes,
                                            title="Save Simulation Results")
    if not filepath:
        return  # User cancelled the save dialog
    # Retrieve simulation data
    t = simulation_results['t']
    Y = simulation_results['Y']  # Expected shape (2, Nt): rows are [S, φ]
    try:
        if file_format == "csv":
            # Prepare columns explicitly: t, S, φ
            data = np.column_stack((t, Y[0, :], Y[1, :]))
            header = "t, S(t), phi(t)"
            np.savetxt(filepath, data, delimiter=",", header=header, comments='')
        elif file_format == "npz":
            np.savez(filepath, t=t, S=Y[0, :], phi=Y[1, :])
        elif file_format == "txt":
            data = np.column_stack((t, Y[0, :], Y[1, :]))
            header = "t, S(t), phi(t)"
            np.savetxt(filepath, data, fmt="%.8e", header=header, comments='')
        messagebox.showinfo("Success", f"Results successfully saved to:\n{filepath}")
    except Exception as e:
        messagebox.showerror("Save Error", f"An error occurred while saving:\n{str(e)}")


def fill_defaults():
    """
    Fill the GUI with default values.
    Φ₀(t) defaults to (N - S0) * A(t).
    """
    for key in ['N', 'S0', 'beta', 'T', 'h']:
        entries[key].delete(0, tk.END)
        entries[key].insert(0, str(DEFAULTS[key]))
    entries['A(t)'].delete(0, tk.END)
    entries['A(t)'].insert(0, DEFAULTS['A(t)'])
    N_val = DEFAULTS['N']
    S0_val = DEFAULTS['S0']
    A_expr = DEFAULTS['A(t)']
    phi0_str = f"({N_val} - {S0_val}) * ({A_expr})"
    entries['Φ₀(t)'].delete(0, tk.END)
    entries['Φ₀(t)'].insert(0, phi0_str)


def close_all():
    """
    Close all open figures and terminate the GUI execution.
    """
    confirm = messagebox.askyesno("Exit Confirmation", "Are you sure you want to close all windows and exit?")
    if confirm:
        plt.close('all')
        root.destroy()


#====================|
# GUI Layout         |
#====================|
root = tk.Tk()
root.attributes('-topmost', True)
root.after(0, lambda: root.attributes('-topmost', False))
root.title("Age-of-Infection Simulation")
frame_inputs = ttk.Frame(root, padding="10")
frame_inputs.grid(row=0, column=0, columnspan=3, sticky=(tk.W, tk.E))
labels = ['N', 'S0', 'beta0', 'T', 'h', 'A(t)', 'Φ₀(t)']
entries = {}
for i, label in enumerate(labels):
    ttk.Label(frame_inputs, text=label).grid(row=i, column=0, sticky=tk.W, pady=2)
    entry = ttk.Entry(frame_inputs, width=30)
    entry.grid(row=i, column=1, pady=2)
    entries[label] = entry
# Control buttons
button_run = ttk.Button(root, text="Run Simulation", command=run_simulation)
button_run.grid(row=1, column=0, pady=10, sticky="ew")
button_defaults = ttk.Button(root, text="Fill in with default values", command=fill_defaults)
button_defaults.grid(row=1, column=1, pady=10, sticky="ew")
# Format selection + Save button
save_format_var = tk.StringVar(value="csv")
format_menu = ttk.OptionMenu(root, save_format_var, "csv", "csv", "npz", "txt")
format_menu.grid(row=1, column=2, padx=5, sticky="ew")
button_save = ttk.Button(root, text="Save Results", command=save_results)
button_save.grid(row=2, column=0, columnspan=2, pady=10, sticky="ew")
# Exit button
button_close = ttk.Button(root, text="Close All Windows", command=close_all)
button_close.grid(row=2, column=2, pady=10, sticky="ew")

#====================|
# Helper Functions   |
#====================|
def safe_float(val):
    return float(val) if val else None

def preprocess_expr(expr):
    return expr.replace('^', '**')

safe_dict = {
    'exp': np.exp, 'log': np.log, 'log10': np.log10, 'sqrt': np.sqrt,
    'sin': np.sin, 'cos': np.cos, 'tan': np.tan,
    'arcsin': np.arcsin, 'arccos': np.arccos, 'arctan': np.arctan,
    'sinh': np.sinh, 'cosh': np.cosh, 'tanh': np.tanh,
    'arcsinh': np.arcsinh, 'arccosh': np.arccosh, 'arctanh': np.arctanh,
    'abs': np.abs, 'sign': np.sign, 'pi': np.pi
}

def make_function(expr):
    if not expr:
        return None
    expr = preprocess_expr(expr)
    return lambda t: np.array(eval(expr, {"__builtins__": None},
                                   {**safe_dict, 't': np.asarray(t, dtype=np.float64), 'np': np}),
                              dtype=np.float64)

def plot_on_ax(ax, t, y, color, title, xlabel='t [days]', ylabel=''):
    y = y if isinstance(y, np.ndarray) else np.array(y)
    ax.plot(t, y, color=color)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(True)

root.mainloop()
```
